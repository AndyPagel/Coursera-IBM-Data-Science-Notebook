```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=10)
```

```python
# Create a blue circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate, radius=1000, color='#0000FF', fill=True).add_child(folium.Popup('NASA Johnson Space Center'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#0000FF;"><b>%s</b></div>' % 'NASA JSC',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)

CCAFS_LC_40_loc = [28.562302,-80.577356]
CCAFS_SLC_40_loc = [28.563197,-80.576820]
KSC_LC_39A_loc = [28.573255,-80.646895]
VAFB_SLC_4E_loc = [34.632834,-120.610745]

site_map_LC = folium.Map(location=CCAFS_LC_40_loc, zoom_start=10)
site_map_SLC = folium.Map(location=CCAFS_SLC_40_loc, zoom_start=10)
site_map_KSC = folium.Map(location=KSC_LC_39A_loc, zoom_start=10)
site_map_VAFB = folium.Map(location=VAFB_SLC_4E_loc, zoom_start=10)

circle_LC = folium.Circle(CCAFS_LC_40_loc, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('CCAFS LC-40'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker_LC = folium.map.Marker(
    CCAFS_LC_40_loc,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'CCAFS LC-40',
        )
    )
site_map.add_child(circle_LC)
site_map.add_child(marker_LC)

circle_SLC = folium.Circle(CCAFS_SLC_40_loc, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('CCAFS SLC-40'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker_SLC = folium.map.Marker(
    CCAFS_SLC_40_loc,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'CCAFS SLC-40',
        )
    )
site_map.add_child(circle_SLC)
site_map.add_child(marker_SLC)


circle_KSC = folium.Circle(KSC_LC_39A_loc, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('KSC LC-39A'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker_KSC = folium.map.Marker(
    KSC_LC_39A_loc,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'KSC LC-39A',
        )
    )
site_map.add_child(circle_KSC)
site_map.add_child(marker_KSC)

circle_VAFB = folium.Circle(VAFB_SLC_4E_loc, radius=1000, color='#d35400', fill=True).add_child(folium.Popup('VAFB SLC-4E'))
# Create a blue circle at NASA Johnson Space Center's coordinate with a icon showing its name
marker_VAFB = folium.map.Marker(
    VAFB_SLC_4E_loc,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % 'VAFB SLC-4E',
        )
    )
site_map.add_child(circle_VAFB)
site_map.add_child(marker_VAFB)
```
